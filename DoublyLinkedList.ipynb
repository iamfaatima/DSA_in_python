```python
class Node:
    def __init__(self, data=None, next=None, prev=None):
        self.data = data
        self.prev = prev
        self.next = next
        
class DoublyLinkedList:
    def __init__(self):
        self.head = None
        
    def printList(self):
        if self.head is None:
            print("empty")
            return
        
        itr = self.head 
        string = ""
        
        while itr:
            string += str(itr.data) 
            if itr.next is not None:
                string += "<-->"
            itr = itr.next
            
        print(string)
        
    def insert_at_end(self, data):
        if self.head is None:
            node = Node(data,None,None)
            self.head = node
            return
        node = Node(data, None, None)
        itr = self.head
        while itr.next:
            itr = itr.next
        node.prev = itr
        itr.next = node 
        
    def delete(self, data):
        if self.head is None:
            return
        if self.head.data == data:
            self.head = self.head.next
            return
        itr = self.head
        while itr.next.data != data:
            itr = itr.next
        if itr.next.next is not None:
            #dont hila its order
            itr.next.next.prev = itr
            itr.next = itr.next.next
        else:
            itr.next = None
            
    def reverse(self):
        #just one pointer temp needed
        if self.head is None:
            return
        if self.head.next == None:
            return
        temp = None
        cur = self.head
        while cur:
            temp = cur.prev
            cur.prev = cur.next
            cur.next = temp
            cur = cur.prev #to iterate
            
        self.head = temp.prev
        #not making self to cur because cur will nikal aagy
        
    def delete_occurences(self, data):
        if self.head is None:
            return
        itr = self.head
        while itr:
            if itr.data == data:
                self.delete(itr.data)
            itr = itr.next
            
    def find_pairs(self, data):
        """
        If current sum of first and second is less than x, 
        then we move first in forward direction. If current 
        sum of first and second element is greater than x, then
        we move second in backward direction.
        """
        if self.head is None:
            return
        itr = self.head
        first = itr
        while itr.next:
            itr = itr.next
        last = itr
        while first != last and first.prev != last:
            #agar main or likhungi tou ek condition true aur ek false hojayegi us 
            #se loop continue rhyga. mene and likha taaky loop break hojaye
            if first.data + last.data == data:
                print("pair: (" + str(first.data) + ", " + str(last.data) + ")")
                first = first.next
                last = last.prev
            elif first.data + last.data < data:
                first = first.next
            elif first.data +last.data > data:
                last = last.prev
                
    def remove_duplicates(self):
        if self.head is None:
            return
        itr = self.head
        s = set()
        while itr.next:
            if itr.data in s:
                self.delete(itr.data)
            s.add(itr.data)
            itr = itr.next
        
if __name__ == "__main__":
    ll = DoublyLinkedList()
    ll.insert_at_end(1)
    ll.insert_at_end(1)
    ll.insert_at_end(2)
    ll.insert_at_end(2)
    ll.insert_at_end(6)
    ll.insert_at_end(8)
    ll.insert_at_end(10)
    ll.printList()
    ll.remove_duplicates()
    ll.printList()
```

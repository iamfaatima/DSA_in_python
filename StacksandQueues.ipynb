```python
#implement stack using arrays
#push, pop, isempty, isfull, peek/top
class Stack:
    def __init__(self, capacity):
        self.capacity = capacity
        self.stack = [None] * capacity
        self.top = -1
        
    def isfull(self):
        if self.top >= self.capacity:
            return True
        else:
            return False

    def isempty(self):
        if self.top == -1:
            return True
        else:
            return False

    def push(self, data):
        if self.top < self.capacity:
            self.top = self.top + 1
            self.stack[self.top] = data
            return data
        else:
            return "stack is full"

    def pop(self):
        if self.top != -1:
            ret = self.stack[self.top]
            self.top = self.top - 1
            return ret
        else:
            return "stack is empty"
        
    def peek(self):
        if self.top != -1:
            return self.stack[self.top]
        else:
            return "stack is empty"
    
    def printStack(self):
        for item in range(self.top+1):
            print(self.stack[item])
        
if __name__ == "__main__":
    s = Stack(55)
    s.push(1)
    s.push(2)
    s.push(3)
    s.push(4)
    s.push(5)
    print(s.peek())
    #print(s.peek())
    s.printStack()
```

```python
#Implement Queue using Arrays
#ismepty, isfull, enqueue, dequeue
class Queue:
    def __init__(self, capacity):
        self.capacity = capacity
        self.queue = [None] * capacity
        self.head = 0
        self.tail = 0
        
    def isfull(self):
        if (self.tail + 1)%self.capacity == self.head:
            return True
        else:
            return False

    def isempty(self):
        if self.head == self.tail:
            return True
        else:
            return False

    def enqueue(self, data):
        if self.isfull():
            return "queue is full"
        else:
            self.queue[self.tail] = data
            self.tail = (self.tail + 1) % self.capacity
            return data

    def dequeue(self):
        if self.isempty():
            return "queue is empty"
        else:
            ret = self.queue[self.head]
            self.queue[self.head] = None
            self.head = (self.head + 1) % self.capacity
            return ret
        
    def peek(self):
        if self.isempty():
            return "queue is empty"
        else:
            ret = self.queue[self.head]
            return ret
    
    
    def printQueue(self):
        for item in range((self.tail - self.head )%self.capacity):
            print(self.queue[item])
            
if __name__ == "__main__":
    q = Queue(55)
    q.enqueue(1)
    q.enqueue(2)
    q.enqueue(3)
    q.enqueue(4)
    q.enqueue(5)
    q.printQueue()
    print("dequeue")
    q.dequeue()
    q.printQueue()
```

```python
#Implement Stack using Queue
#push, pop, isempty, isfull, peek/top
class Stack2:
    def __init__(self, capacity):
        self.capacity = capacity
        self.queue1 = Queue(55)
        self.queue2 = Queue(55)
        
    def isfull(self):
        if self.queue1.isfull() and self.queue2.isfull():
            return True
        else:
            return False

    def isempty(self):
        if self.queue1.isempty() and self.queue2.isempty:
            return True
        else:
            return False

    def push(self, data):
        if self.queue1.isfull():
            if self.queue2.isfull():
                return None
            self.queue2.enqueue(data)
            while not self.queue1.isempty():
                self.queue2.enqueue(self.queue1.dequeue())
            self.queue1, self.queue2 = self.queue2, self.queue1
        else:
            self.queue1.enqueue(data)
            while not self.queue2.isempty():
                self.queue1.enqueue(self.queue2.dequeue())
            self.queue1, self.queue2 = self.queue2, self.queue1

    def pop(self):
        if not self.queue1.isempty():
            self.queue1.pop()
        else:
            return "stack is empty"
        
    def peek(self):
        if self.queue1.isempty():
            return None
        return self.queue1.peek()
    
    def printStack(self):
        self.queue1.printQueue()
        
if __name__ == "__main__":
    s = Stack2(55)
    s.push(1)
    s.push(1)
    s.push(1)
    s.printStack()
    #print(s.peek())
    #s.printStack()
```
